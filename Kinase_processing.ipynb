# **Data Processing of Kinase Dataset**

# Load Dataset

```python
!gdown --id 1MhgGWkQ4lB7IjBXeF2Ohv0CcmnEDTmw5
```

```python
import pandas as pd

xls = pd.ExcelFile('Kinase-All_ChEMBL+pref_name.xlsx')
df = pd.read_excel(xls, 'Kinase-All_ChEMBL_220921')
df
```

# Save Dataset to CSV

```python
df.to_csv('Kinase-All_ChEMBL+pref_name.csv')
```

# Group Data

```python
target_molecule_group = df.groupby(['Target ChEMBL ID', 'Molecule ChEMBL ID', 'Smiles', 'Type_MPG'])
```

```python
df3 = target_molecule_group['pChEMBL Value'].agg(['min', 'max', 'mean', 'std'])
df3
```

```python
df3.to_csv('Kinase-All_ChEMBL_Group.csv')
```

# **Select only Cell from Type_MPG**

```python
df3 = df.loc[df['Type_MPG'] == 'CELL']
df3
```

# Replace NaN = 0

```python
df4 = df3.fillna(0)
df4
```

# Save Dataset to CSV

```python
df4.to_csv('Kinase-All_ChEMBL_TypeMPG_CELL.csv')
```

# **Select only Protein from Type_MPG**

```python
df5 = df.loc[df['Type_MPG'] == 'PROTEIN']
df5
```

# Replace NaN = 0

```python
df6 = df5.fillna(0)
df6
```

# Save Dataset to CSV

```python
df6.to_csv('Kinase-All_ChEMBL_TypeMPG_PROTEIN.csv')
```

# **Select only Unclear from Type_MPG**

```python
df7 = df.loc[df['Type_MPG'] == 'UNCLEAR']
df7
```

# Replace NaN = 0

```python
df8 = df7.fillna(0)
df8
```

# Save Dataset to CSV

```python
df8.to_csv('Kinase-All_ChEMBL_TypeMPG_UNCLEAR.csv')
```
